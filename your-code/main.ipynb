## Distribución Uniforme

Para generar números aleatorios uniformes entre dos valores dados usando scipy, podemos usar el siguiente código o el código que hemos
discutido en clase:

```python
# your code here
def generar_uniforme(bottom, ceiling, count):
    return np.random.uniform(bottom, ceiling, count)

# Parámetros
bottom1, ceiling1, count1 = 10, 15, 100
bottom2, ceiling2, count2 = 10, 60, 1000

# Generar números
data1 = generar_uniforme(bottom1, ceiling1, count1)
data2 = generar_uniforme(bottom2, ceiling2, count2)

# Graficar histogramas
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data1, bins=10)
plt.subplot(1, 2, 2)
plt.hist(data2, bins=10)
plt.show()
```

## Distribución Exponencial

1. Utilizando `numpy.random.exponential`, crea una función que devuelva una lista de números distribuidos exponencialmente con la media de 10. 

1. Utiliza la función para generar dos secuencias de números de tamaño 10 y 100.

1. 2. Traza las distribuciones como histogramas con el número de bins como 100.

El resultado debe ser como el siguiente:



```python
import numpy as np

# Crear una función que devuelva una lista de números distribuidos exponencialmente con la media de 10
def generar_exponencial(media, count):
    return np.random.exponential(media, count)

# Generar una secuencia de números distribuidos exponencialmente con tamaño 10
data1 = generar_exponencial(10, 10)
print(data1)

# Generar una secuencia de números distribuidos exponencialmente con tamaño 100
data2 = generar_exponencial(10, 100)
print(data2)
```

```python
import matplotlib.pyplot as plt

# Trazar el histograma para la secuencia de tamaño 10
plt.figure(figsize=(6, 6))
plt.hist(data1, bins=100)
plt.title('Distribución Exponencial (n=10)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Trazar el histograma para la secuencia de tamaño 100
plt.figure(figsize=(6, 6))
plt.hist(data2, bins=100)
plt.title('Distribución Exponencial (n=100)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

### ¿En qué se diferencian ambas distribuciones?

La primera distribución (tamaño 10) es mucho más ruidosa debido al pequeño tamaño de la muestra. La segunda distribución (tamaño 100) es más suave y se parece más a la distribución exponencial teórica debido al mayor tamaño de la muestra.

La media cambia, por lo que la distribución también cambia. 

## Distribución Exponencial de datos reales

Supongamos que el tiempo que uno pasa en un banco se distribuye exponencialmente con media de 10 minutos (es decir, λ = 1/10). ¿Cuál es la probabilidad de que un cliente pase menos de quince minutos en el banco? 

Escribe un código en python para resolver este problema

```python
# Parámetros
media = 10
lambd = 1 / media

# Probabilidad de pasar menos de 15 minutos
probabilidad_menos_15 = 1 - np.exp(-lambd * 15)
probabilidad_menos_15
```

¿Cuál es la probabilidad de que el cliente pase más de 15 minutos?
